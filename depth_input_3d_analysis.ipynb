```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
def plot3d(img, x_rot=50, z_rot=30):
    xx, yy = np.mgrid[0 : img.shape[0], 0 : img.shape[1]]

    fig = plt.figure(figsize=(20, 20))
    ax = fig.gca(projection='3d')

    ax.plot_surface(xx, yy, img, rstride=1, cstride=1, cmap=plt.cm.gray, linewidth=9)
    ax.view_init(x_rot, z_rot)

    plt.show()
```

```python
data_path = 'depth_roi_data/test_depth_img.png'
depth = cv2.imread(data_path, cv2.IMREAD_ANYDEPTH).astype(np.uint16)
plt.imshow(depth, cmap='gray')
```

```python
depth_norm = cv2.normalize(depth, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
color = cv2.applyColorMap(depth_norm, cv2.COLORMAP_JET)
plot3d(depth)
```
