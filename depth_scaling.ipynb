```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

from skimage import data, color
from skimage.transform import rescale, resize, downscale_local_mean

%matplotlib inline
```

```python
depth_roi = cv2.imread('depth_roi.png', cv2.IMREAD_ANYDEPTH)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1)

#HWC
depth_tmp = depth_roi[47 : 58, 36 : 46]
ax.imshow(depth_tmp, cmap='gray')

rows = 11
cols = 10

for row in np.arange(0, depth_tmp.shape[0]):
    for col in np.arange(0, depth_tmp.shape[1]):
        ax.annotate(
            str(depth_roi[row, col]),
            xy=(col, row),
            horizontalalignment='center',
            verticalalignment='center',
            color='gray' if depth_roi[row, col] == 0 else 'white'
        )
```

```python
downscaled = rescale(depth_roi, 0.25, anti_aliasing=True)

fig = plt.figure(figsize=(15, 15))
ax1 = fig.add_subplot(1, 2, 1)
ax1.set_title('Downscaled')
ax1.imshow(downscaled, cmap='gray')

ax2 = fig.add_subplot(1, 2, 2)
ax2.set_title('Original')
ax2.imshow(depth_roi, cmap='gray')
```

```python
upscaled = rescale(depth_roi, 1.25, anti_aliasing=False)

fig = plt.figure(figsize=(15, 15))
ax1 = fig.add_subplot(1, 2, 1)
ax1.set_title('Upscaled')
ax1.imshow(upscaled, cmap='gray')

ax2 = fig.add_subplot(1, 2, 2)
ax2.set_title('Original')
ax2.imshow(depth_roi, cmap='gray')
```

```python
downsized = resize(depth_roi, (224, 224))

fig = plt.figure(figsize=(15, 15))
ax1 = fig.add_subplot(1, 2, 1)
ax1.set_title('Downsized')
ax1.imshow(downsized, cmap='gray')

ax2 = fig.add_subplot(1, 2, 2)
ax2.set_title('Original')
ax2.imshow(depth_roi, cmap='gray')
```

```python
upsized = resize(depth_roi, (500, 500))

fig = plt.figure(figsize=(15, 15))
ax1 = fig.add_subplot(1, 2, 1)
ax1.set_title('Upsized')
ax1.imshow(upsized, cmap='gray')

ax2 = fig.add_subplot(1, 2, 2)
ax2.set_title('Original')
ax2.imshow(depth_roi, cmap='gray')
```

```python
downscaled_mean = downscale_local_mean(depth_roi, (100, 100))

fig = plt.figure(figsize=(15, 15))
ax1 = fig.add_subplot(1, 2, 1)
ax1.set_title('Downscaled_mean')
ax1.imshow(downscaled_mean, cmap='gray')

ax2 = fig.add_subplot(1, 2, 2)
ax2.set_title('Original')
ax2.imshow(depth_roi, cmap='gray')
```

```python
cv_resized = cv2.resize(depth_roi, (100, 100), cv2.INTER_AREA)

fig = plt.figure(figsize=(15, 15))
ax1 = fig.add_subplot(1, 2, 1)
ax1.set_title('cv2 Resized')
ax1.imshow(cv_resized, cmap='gray')

ax2 = fig.add_subplot(1, 2, 2)
ax2.set_title('Original')
ax2.imshow(depth_roi, cmap='gray')
```
