```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

from mpl_toolkits import mplot3d

%matplotlib notebook
```

```python
sharpen = np.array([
    [ -1, -1, -1 ],
    [ -1, 9, -1 ],
    [ -1, -1, -1 ]
])

smooth = 1 / 16 * np.array([
    [ 1, 2, 2 ],
    [ 2, 4, 2 ],
    [ 1, 2, 1 ]
])

def plot3d(img, x_rot=30, z_rot=10):
    xx, yy = np.mgrid[0 : img.shape[0], 0 : img.shape[1]]

    fig = plt.figure(figsize=(20, 20))
    ax = fig.gca(projection='3d')

    ax.plot_surface(xx, yy, img, rstride=1, cstride=1, cmap=plt.cm.gray, linewidth=9)
    ax.view_init(x_rot, z_rot)

    plt.show()
```

```python
depth = cv2.imread('depth_roi_data/spoof_depth_roi2.png', cv2.IMREAD_ANYDEPTH).astype(np.uint16)
depth = cv2.filter2D(depth, -1, smooth)
plt.imshow(depth, cmap='gray')

plot3d(depth)
```

```python
chroma = cv2.cvtColor(cv2.imread('depth_roi_data/spoof_color_roi2.png').astype(np.uint8), cv2.COLOR_BGR2YCrCb)[:, :, 1]
plt.imshow(chroma, cmap='gray')

plot3d(chroma, 80)
```

```python
depth = cv2.imread('depth_roi_data/real_depth_roi2.png', cv2.IMREAD_ANYDEPTH).astype(np.uint8)
depth = cv2.filter2D(depth, -1, smooth)
plt.imshow(depth, cmap='gray')

plot3d(depth)
```
